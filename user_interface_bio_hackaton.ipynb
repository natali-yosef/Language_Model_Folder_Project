```python
#@title imports

from IPython import display
import ipywidgets as widgets
import numpy as np
import sys
from google.colab import files
from IPython.display import display

%cd /content/Language_Model_Folder_Project/

import py3Dmol
import utils
import net
```

```python
#@title Prediction 3D Viewer

def show_prediction(predicted_pdb,view):
  view.removeAllModels()

  # Load the PDB file into the viewer
  view.addModel(open(predicted_pdb, 'r').read(), 'pdb')
  view.setStyle({'cartoon': {'color': 'spectrum'}})

  # Set the background color
  view.setBackgroundColor('white')

  # Center and zoom the view
  view.zoomTo()
  # Show the viewer
  view.update()
```

```python
#@title get protein sequence from user

protein_sequence_name = input("Enter protein name for prediction: ")
protein_sequence = input("Enter protein sequence for prediction: ")
embeding_dropdown = widgets.Dropdown(options= utils.MODELS_LIST, description='Select an embeding option:',style={'description_width': 'initial'})

predict_button = widgets.Button(description="Predict!",style={'description_width': 'initial'})
predicted_pdb = None

view = py3Dmol.view(width=800, height=600)

# Set the style for the protein structure
view.setStyle({'cartoon': {'color': 'spectrum'}})

# Set the background color
view.setBackgroundColor('white')

# Center and zoom the view
view.zoomTo()


def start_predict(b):
    embeding_dropdown_value = embeding_dropdown.value
    show_prediction(net.use_model(protein_sequence, protein_sequence_name, embeding_dropdown_value), view)

predict_button.on_click(start_predict)

display(embeding_dropdown, predict_button)

view.show()
```
